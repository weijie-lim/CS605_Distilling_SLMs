# SLM trained on 25% of LLM-generated Labels

```python
import json

with open('result_25_percent_trained.jsonl', 'r') as json_file:
  json_list = list(json_file)

# with open('SLM_inputs_with_prompts_train.jsonl', 'w') as output_file:
l = []
for json_str in json_list:
  result = json.loads(json_str)
  l.append(result)
print(l[0]) 
```

```python
pred = []
ans = []
for i in l:
  ans.append(i['actual_answer'])
  
  output = i['output'][0][0].split(" ")[-1].replace(".", "")
  if output.isnumeric():
    pass
  else:
    output = i['output'][0][0].split(" ")[-2].replace(".", "")
    if not output.isnumeric():
      output = None
  
  pred.append(output)

correct = 0
for i, j in zip(pred, ans):
  if i == j:
    correct += 1
  
print(pred)
print(ans)
print(correct / len(pred))
```
